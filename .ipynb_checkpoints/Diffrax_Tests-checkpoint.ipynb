## Neural ODEs:
$$ y(t) = y(0) + \int_0^t f_\theta(s, y(s))ds$$

```python
class Func(eqx.Module):
    mlp: eqx.nn.MLP

    def __init__(self, data_size, width_size, depth, *, key, **kwargs):
        super().__init__(**kwargs)
        self.mlp = eqx.nn.MLP(
            in_size=data_size,
            out_size=data_size,
            width_size=width_size,
            depth=depth,
            activation=jnn.softplus,
            key=key,
        )

    def __call__(self, t, y, args):
        return self.mlp(y)
```

### Coordinates:

```python
Nx, Ny, Nt = 64, 64, 256
Lx, Ly, T = 1.0, 1.0, 1.0
dx = Lx/Nx
dy = Ly/Ny
dt = T / Nt

x = jnp.linspace(-Lx/2, Lx/2, Nx)
y = jnp.linspace(-Ly/2, Ly/2, Ny)
t = jnp.linspace(-T/2, T/2, Nt)
X, Y, T_grid = jnp.meshgrid(x,y,t, indexing = 'ij')

kx= 2* jnp.pi * jnp.fft.fftfreq(Nx, d=dx)
ky= 2*jnp.pi * jnp.fft.fftfreq(Ny, d=dy)
omega = 2*jnp.pi*jnp.fft.fftfreq(Nt, d=dt)
KX, KY, OMEGA = jnp.meshgrid(kx, ky, omega, indexing = 'ij')
```

### Physical Parameters

```python
k0 = 2*jnp.pi/ 1.55e-6  # central wavenumber
kpp = 1e-26 # group velocity dispersion
gamma = 1.0 # nonlinear coefficient
```

```python
def linear_step(A, dz):
    A_fft = jnp.fft.fftn(A)

    phase_linear = jnp.exp(
        1j * dz *( -(KX**2 + KY**2)/(2*k0)  + (kpp/2) * (OMEGA**2))
    )
    A_fft_new = A_fft * phase_linear
    A_new = jnp.fft.ifftn(A_fft_new)
    
    return A_new


def nonlinear_step(A, dz):

    phase_nonlinear = jnp.exp(1j * dz * gamma * jnp.abs(A)**2)
    
    return A * phase_nonlinear
    

def split_step(A, dz):
    
    A1 = nonlinear_step(A, dz/2)
    A2 = linear_step(A1, dz)
    
    return nonlinear_step(A2, dz/2)
    

def propagate_field(A0, dz, num_steps):

    A=A0
    
    for _ in tqdm.tqdm(range(num_steps)):
        A = split_step(A, dz)
        
    return A


    
```

```python
def gnlse_rhs(z, A_flat, args):
    A = A_flat.reshape((Nx, Ny, Nt))

    A_fft = jnp.fft.fftn(A)
    phase_factor = 1j * ( -(KX**2 + KY**2)/(2*k0)  + (kpp/2) * (OMEGA**2))
    A_linear = jnp.fft.ifftn(A_fft * phase_factor)

    A_nonlinear = 1j * gamma * jnp.abs(A)**2 * A

    dA_dz = A_linear + A_nonlinear
    return dA_dz.ravel()

# We choose an initial field with a Gaussian spatial beam profile and a Gaussian temporal pulse.
beam_waist = 0.1
pulse_duration = 0.1
A0 = jnp.exp(-((X**2 + Y**2) / (beam_waist**2))) * jnp.exp(- (T_grid**2) / (pulse_duration**2))

# <-- Fix: Convert A0 to a complex type.
A0 = A0.astype(jnp.complex64)  # or jnp.complex128 depending on your precision needs

# Flatten the initial condition for diffrax:
A0_flat = A0.ravel()


ode_term = diffrax.ODETerm(gnlse_rhs)
params = {
    "k0": k0,
    "kpp": kpp,
    "gamma": gamma,
    "KX": KX,
    "KY": KY,
    "OMEGA": OMEGA,
    "Nx": Nx,
    "Ny": Ny,
    "Nt": Nt,
}
z0 = 0.0
z1 = 1.0
dt0 = 0.001
solver = diffrax.Dopri5()
num_z_samples = 50
ts = jnp.linspace(z0, z1, num_z_samples)
sol = diffrax.diffeqsolve(ode_term, solver, t0 = z0, t1=z1, dt0=dt0, y0=A0_flat, args=params, 
                          saveat=diffrax.SaveAt(ts=ts))

A_final = sol.ys[-1].reshape((Nx, Ny, Nt))
A_final_np = np.array(A_final)
I_final = np.abs(A_final_np)**2
```

```python
# (a) Final propagated field: reshape the final field to 3D.
A_final = sol.ys[-1].reshape((Nx, Ny, Nt))
# Convert the JAX array to a NumPy array for plotting:
A_final_np = np.array(A_final)
# Compute the intensity, I = |A|^2:
I_final = np.abs(A_final_np)**2

# Plot 1: Intensity distribution in (x,y) at the central time slice (t = 0)
t_index = Nt // 2  # choose the central t slice
y_index = Ny // 2
plt.figure(figsize=(8, 6))
plt.imshow(
    I_final[:, :, t_index],
    extent=[-Ly/2, Ly/2, -Lx/2, Lx/2],
    origin="lower",
    aspect="auto"
)
plt.xlabel('y')
plt.ylabel('x')
plt.title('Intensity in the (x,y) plane at t = 0 (final z)')
plt.colorbar(label='Intensity')
plt.show()

# Plot 2: Cross-sectional intensity aly_index = Ny // 2  # central y value
plt.figure(figsize=(8,6))
plt.plot(x, I_final[:, y_index, t_index])
plt.xlabel('x')
plt.ylabel('Intensity')
plt.title('Intensity along x (at y = 0, t = 0, final z)')
plt.grid(True)
plt.show()

# Plot 3: Evolution of the central intensity (at x=0, y=0, t=0) versus z.
# (We extract the intensity at the center from each saved step.)
central_intensity = []
for sol_y in sol.ys:
    A_current = sol_y.reshape((Nx, Ny, Nt))
    intensity_center = np.abs(A_current[Nx//2, Ny//2, Nt//2])**2
    central_intensity.append(intensity_center)
central_intensity = np.array(central_intensity)

plt.figure(figsize=(8,6))
plt.plot(np.array(ts), central_intensity, marker='o')
plt.xlabel('Propagation Distance z')
plt.ylabel('Central Intensity')
plt.title('Evolution of Central Intensity vs. z')
plt.grid(True)
plt.show()
```

```python
import jax
import jax.numpy as jnp
import diffrax
import numpy as np
import matplotlib.pyplot as plt

#############################
# 1. Define the Spatial and Temporal Grids
#############################
# Grid parameters (number of points)
Nx, Ny, Nt = 128, 128, 256
Lx, Ly = 0.01, 0.01   # Physical transverse size in meters (adjust as needed)
T_win = 5e-12         # Temporal window in seconds

dx = Lx / Nx
dy = Ly / Ny
dt = T_win / Nt

# Create coordinate arrays (centered grid)
x = jnp.linspace(-Lx/2, Lx/2, Nx)
x=jnp.ravel(x)
y = jnp.linspace(-Ly/2, Ly/2, Ny)
y=jnp.ravel(y)
t = jnp.linspace(-T_win/2, T_win/2, Nt)
t=jnp.ravel(t)
# Ensure x and y are one-dimensional:
x = jnp.asarray(x).reshape(-1)  # Force shape (128,)
y = jnp.asarray(y).reshape(-1)  # Force shape (128,)

# Create meshgrids for potential functions (for x and y)
X, Y = jnp.meshgrid(x, y, indexing="ij")
R2 = X**2 + Y**2

#############################
# 2. Define Fourier-Domain Variables
#############################
# Transverse frequency variables (for x and y)
kx = 2 * jnp.pi * jnp.fft.fftfreq(Nx, d=dx)
ky = 2 * jnp.pi * jnp.fft.fftfreq(Ny, d=dy)
KX, KY = jnp.meshgrid(kx, ky, indexing="ij")
laplacian_operator = -(KX**2 + KY**2)  # multiplier in Fourier space for ∇² in (x,y)

# Temporal frequency variable (for t)
omega = 2 * jnp.pi * jnp.fft.fftfreq(Nt, d=dt)

#############################
# 3. Physical Parameters for the Fiber
#############################
# For demonstration, we set:
wavelength = 1.55e-6         # Central wavelength (m)
omega0 = 2*jnp.pi*3e8/wavelength  # Central angular frequency
n0 = 1.45                    # Average refractive index
beta0 = n0 * omega0 / 3e8    # Approximate propagation constant (1/m)
beta2 = 20e-27               # Group velocity dispersion parameter (s^2/m) -- adjust as needed

# For a GRIN fiber, the refractive index profile is quadratic. The effective potential is often modeled as
kappa = 1e8  # GRIN strength (1/m^3) [adjust to set focusing strength]
# Define potential V(x,y): parabolic for GRIN fiber.
V = 0.5 * kappa * R2  # Units of 1/m (when scaled appropriately)

# For a step-index fiber, one could define V as:
# core_radius = 0.002  # in meters
# V = jnp.where(R2 <= core_radius**2, 0.0, 1e5)

# Spatially dependent nonlinear coefficient (units: 1/(W*m))
# For simplicity, assume it is constant in the core and zero outside,
# or use a smoothly varying function.
gamma = 1e-3 * jnp.ones((Nx, Ny))  # adjust as needed (1/(W*m))

# Define a PML (Perfectly Matched Layer) absorption profile W(x,y)
# Let the PML start at a distance pml_start from the edge and increase smoothly.
pml_thickness = 10  # number of grid points for PML in each direction
W_max = 1e4  # maximum absorption (1/m) at the boundaries

def pml_profile(coord, pml_thickness, size):
    # Create a 1D PML profile that is 0 in the middle and increases smoothly near the edges.
    profile = jnp.zeros(coord.shape)
    for i in range(coord.shape[0]):
        # Compute the distance to the nearest boundary (in grid points)
        dist = jnp.minimum(i, coord.shape[0]-i-1)
        profile = profile.at[i].set(jnp.where(dist < pml_thickness,
                                                W_max * ((pml_thickness - dist)/pml_thickness)**2,
                                                0.0))
    return profile

W_x = jnp.asarray(pml_profile(x, pml_thickness, Lx)).reshape(-1)
W_y = jnp.asarray(pml_profile(y, pml_thickness, Ly)).reshape(-1)

# Form a 2D PML profile by outer addition; then broadcast to the time dimension later.
W = W_x[:, None] + W_y[None, :]

#W = jnp.squeeze(W, axis=0)  # Force shape to (128, 128) if an extra dimension appears


#############################
# 4. Define the ODE Right-Hand Side for the 3D GNLSE
#############################
def gnlse_rhs(z, A_flat, args):
    """
    Right-hand side function for the full-field (x,y,t) evolution.
    
    The equation is:
      ∂z A = i [ (1/(2β₀)) ∇⊥² A  - (β₂/2) ∂²ₜ A  + V(x,y) A + γ(x,y)|A|²A ]
              + W(x,y) A.
              
    FFT-based derivatives are used along (x,y) for the Laplacian
    and along t for the second derivative.
    """
    # Unpack parameters
    Nx = args["Nx"]
    Ny = args["Ny"]
    Nt = args["Nt"]
    beta0 = args["beta0"]
    beta2 = args["beta2"]
    lap_op = args["lap_op"]  # (Nx, Ny) Laplacian multiplier in Fourier space (for x,y)
    omega = args["omega"]    # (Nt,) temporal frequencies
    V = args["V"]            # potential (Nx,Ny)
    gamma_space = args["gamma"]  # nonlinear coefficient (Nx,Ny)
    W = args["W"]            # PML absorption (Nx,Ny)
    
    # Reshape the state into (Nx,Ny,Nt)
    A = A_flat.reshape((Nx, Ny, Nt))
    
    # --- Diffraction (Transverse Laplacian) ---
    # FFT along x and y (axes 0 and 1)
    A_xy_fft = jnp.fft.fftn(A, axes=(0,1))
    laplacian_xy = jnp.fft.ifftn(A_xy_fft * lap_op, axes=(0,1))
    term_diffraction = (1/(2*beta0)) * laplacian_xy

    # --- Dispersion (Temporal Second Derivative) ---
    # FFT along t (axis 2)
    A_t_fft = jnp.fft.fft(A, axis=2)
    # Multiply by -omega^2 (note: omega is 1D; use broadcasting over (Nx,Ny))
    # Here we use: ∂²A/∂t² = IFFT( -ω² FFT(A) )
    d2A_dt2 = jnp.fft.ifft(A_t_fft * (- (omega**2))[None, None, :], axis=2)
    term_dispersion = - (beta2/2) * d2A_dt2  # note the minus sign from the equation

    # --- Waveguiding Potential ---
    # V is a function of (x,y); broadcast it along t.
    term_potential = V[:, :, None] * A

    # --- Nonlinear Term ---
    nonlinear_phase = gamma_space[:, :, None] * jnp.abs(A)**2
    term_nonlinear = nonlinear_phase * A

    # --- Combine Linear Terms with factor i and add PML term ---
    # The PML term is added without the factor i.
    RHS = 1j * (term_diffraction + term_dispersion + term_potential + term_nonlinear) + W[:, :, None] * A
    
    return RHS.ravel()

#############################
# 5. Assemble Parameters and Initial Condition
#############################
args = {
    "Nx": Nx,
    "Ny": Ny,
    "Nt": Nt,
    "beta0": beta0,
    "beta2": beta2,
    "lap_op": laplacian_operator,  # shape (Nx,Ny)
    "omega": omega,                # shape (Nt,)
    "V": V,                        # shape (Nx,Ny)
    "gamma": gamma,                # shape (Nx,Ny)
    "W": W,                        # shape (Nx,Ny)
}

# Define an initial condition A(x,y,t,0)
# For example, a Gaussian beam (in x,y) combined with a short Gaussian pulse (in t):
beam_waist = 0.002   # meters
pulse_duration = 0.5e-12  # seconds
X3d, Y3d, T3d = jnp.meshgrid(x, y, t, indexing="ij")
A0 = jnp.exp(-((X3d**2 + Y3d**2) / (beam_waist**2))) * jnp.exp(- (T3d**2) / (pulse_duration**2))
# Ensure the field is complex
A0 = A0.astype(jnp.complex64)
A0_flat = A0.ravel()

#############################
# 6. Solve the ODE with diffrax
#############################
# We use diffrax's ODETerm to wrap the RHS.
term = diffrax.ODETerm(gnlse_rhs)
z0 = 0.0      # Starting propagation distance (m)
z1 = 0.05     # Final propagation distance (m)
dt0 = 1e-5    # Initial step size along z

# Choose an ODE solver (we use Dopri5 here)
solver = diffrax.Dopri5()

# Optionally, save the solution at specified z values.
num_z_samples = 20
zs = jnp.linspace(z0, z1, num_z_samples)

sol = diffrax.diffeqsolve(
    term,
    solver,
    t0=z0,
    t1=z1,
    dt0=dt0,
    y0=A0_flat,
    args=args,
    saveat=diffrax.SaveAt(ts=zs)
)

#############################
# 7. Visualization
#############################
# For illustration, we extract the final field and a transverse cross-section.
A_final = sol.ys[-1].reshape((Nx, Ny, Nt))
I_final = jnp.abs(A_final)**2  # Intensity

# Plot (x,y) intensity at the temporal center (t ~ 0)
t_index = Nt // 2
plt.figure(figsize=(8, 6))
plt.imshow(np.array(I_final[:, :, t_index]),
           extent=[y[0], y[-1], x[0], x[-1]],
           origin="lower", aspect="auto")
plt.xlabel("y [m]")
plt.ylabel("x [m]")
plt.title("Transverse Intensity at t=0, z = {:.3f} m".format(z1))
plt.colorbar(label="Intensity (a.u.)")
plt.show()

# Plot temporal intensity at the fiber center (x=0, y=0)
x_index = Nx // 2
y_index = Ny // 2
plt.figure(figsize=(8, 6))
plt.plot(t, np.array(I_final[x_index, y_index, :]), "o-")
plt.xlabel("Time [s]")
plt.ylabel("Intensity (a.u.)")
plt.title("Temporal Intensity at x=0, y=0, z = {:.3f} m".format(z1))
plt.grid(True)
plt.show()
```
